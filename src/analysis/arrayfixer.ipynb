```python
import csv
```

```python
def heal_the_csv(filename,finalfilename):
    healed_array = []
    with open(filename, 'r') as newfile:
        fobject = list(csv.reader(newfile))
        header = fobject.pop(0)
        datapoint = ""
        for i in fobject:
            datapoints = []
            for j in i:
                if j==',':
                    datapoints.append(datapoint)
                    datapoint=""
                elif j=='\n':
                    datapoints.append(datapoint)
                    datapoint=""
                    break
                else:
                    datapoint+=j
            healed_array.append(datapoints)
    with open(finalfilename,'a',newline="") as datafile:
        writer = csv.writer(datafile)
        writer.writerow(header)
        writer.writerows(healed_array)
```

```python
heal_the_csv("../data/background2.csv","../data/healedbackground2.csv")
```
